```python
print('City in which most matches won:', matches['city'].value_counts().idxmax())
print('which team has won most of the matches', matches['winner'].value_counts().idxmax())
print("Player who has been man of the match most times: ",matches['player_of_match'].value_counts().idxmax()) 
print('Most frequent Umpire 1', matches['umpire1'].value_counts().idxmax())
print('Most frequent Umpire 2', matches['umpire2'].value_counts().idxmax())
```

We are going to replace the missing values with the above outputs for their respective columns.  
Since most values are null in umpire 3 we will replace them by NA.

```python
matches['city'].fillna(value='Mumbai', inplace=True)
matches['winner'].fillna(value='Mumbai Indians', inplace=True)
matches['player_of_match'].fillna(value='CH Gayle', inplace=True)
matches['umpire1'].fillna(value='HDPK Dharmasena', inplace=True)
matches['umpire2'].fillna(value='C Shamshuddin', inplace=True)
matches['umpire3'].fillna(value='NA', inplace=True)
```

```python
matches.isnull().sum()
```

```python
matches.duplicated().sum()
```

Now we can see there is no such duplicates and null values.

**Merging the two Datasets into a new Dataset and Reading it (join on match-id)**

```python
delivery = pd.merge(deliveries, matches, left_on='match_id', right_on='id')
delivery.head()
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(delivery.corr(), annot=True, linewidth=1, cmap='twilight_shifted')
plt.show()
```

# **Number of matches in each stadium**

```python
plt.figure(figsize=(16,16))
delivery.venue.value_counts()
sns.countplot(data=delivery, y='venue')
plt.title('Number of matches in each stadium', fontsize=15)
plt.xlabel('Count', fontsize=15)
plt.ylabel('Venue/Stadium', fontsize=15)
plt.show()
```

Most matches have been played in Eden Gardens followed by Wankhede Stadium and M Chinnaswamy Stadium.

```python
team_stats = pd.DataFrame({'Total match Played' : matches.team1.value_counts() + matches.team2.value_counts(),
                          'Total Won' : matches.winner.value_counts(),
                          'Toss Won' : matches.toss_winner.value_counts(),
                          'Total Loss' : matches.team1.value_counts() + matches.team2.value_counts() - matches.winner.value_counts()})

team_stats = team_stats.reset_index()
team_stats.rename(columns = {'index' : 'Teams'}, inplace=True)
winloss = team_stats['Total Won'] / team_stats['Total match Played']
winloss = pd.DataFrame({'Winloss team' : team_stats['Total Won'] / team_stats['Total match Played']})
winloss = winloss.round(2)
team_stats = team_stats.join(winloss)
team_stats
```

# Deciding Whether to Bat or Field After Winning the Toss 

```python
plt.figure(figsize=(16,8))
matches['toss_decision'].value_counts().plot(kind='pie', fontsize=16, autopct='%.2f%%', legend=True)
plt.title('Deciding Whether to Bat or Field After Winning the Toss', fontsize=15)
plt.show()
```

Around 61% times teams who have won tosses have decided to chase down

# Relation between Winning toss and victory

```python
delivery['team_toss_win']=np.where((delivery.toss_winner == delivery.winner),'Yes','No')
plt.figure(figsize=(16,8))
sns.countplot(delivery['team_toss_win'], data=delivery, hue='toss_decision', palette='rocket_r')
plt.xlabel("Winning the Toss Vs Winning the Match", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.title("Toss Wins Vs Victory", fontsize=15)
plt.show()
```

Teams who choose to field after winning the toss have high chances of winning.

```python
plt.figure(figsize=(25,20))
sns.countplot(matches['toss_winner'], data=matches, hue='toss_decision', palette='viridis')
plt.title('Teams decision to bat first or second after winning toss', fontsize=20, fontweight='bold')
plt.xticks(rotation=50, size=15)
plt.yticks(size=15)
plt.xlabel('Toss Winner Team', size=20)
plt.legend()
plt.ylabel('Count', size=20)
plt.show()
```

Most teams field first after winning toss except for Chennai Super Kings who has mostly opted to bat first. Deccan Chargers and Pune Warriors also show the same trend.

# Batsmen overview

```python
batsman = delivery.groupby("batsman").agg({'ball': 'count','batsman_runs': 'sum'})
batsman.rename(columns={'ball':'balls', 'batsman_runs': 'runs'}, inplace=True)
batsman = batsman.sort_values(['balls','runs'], ascending=False)
batsman['batting_strike_rate'] = batsman['runs']/batsman['balls'] * 100
batsman['batting_strike_rate'] = batsman['batting_strike_rate'].round(2)
batsman.head(10)
```

```python
TopBatsman = batsman.sort_values(['balls','runs'], ascending=False)[:20]
TopBatsman
```

# Individual teams decision to choose bat or field after winning toss.

Most teams field first after winning toss except for Chennai Super Kings who has mostly opted to bat first. Deccan Chargers and Pune Warriors also show the same trend.

```python
plt.figure(figsize=(16,8))
delivery.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='orange')
plt.title("Top 10 most Scored Batsman", fontsize=15, fontweight='bold')
plt.xticks(rotation=45, fontsize=15, color='red', fontweight='bold')
plt.yticks(fontsize=15)
plt.xlabel('Batsman', fontsize=15)
plt.ylabel('Runs', fontsize=15)
plt.show()
```

Except for MS Dhoni, all other top run getters are either openers or come in 3rd or 4th positions to bat.  
Virat Kohli is the highest run getter in IPL followed by Suresh Raina and Rohit Sharma.

# Top Wicket Takers Ballers of IPL

```python
plt.figure(figsize=(16,8))
deliveries.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).plot(kind='bar', color='purple')
plt.title('Top 10 most Wickets Takers Ballers', fontsize=15, fontweight='bold')
plt.xticks(rotation=50, fontsize=15, color='black')
plt.yticks(fontsize=15, color='black')
plt.xlabel('Bowler', fontsize=15)
plt.ylabel('Wickets', fontsize=15)
plt.show()
```

Malinga has taken the most number of wickets in IPL followed by Bravo and Amit Mishra  
In top 10 bowlers, 5 are fast and medium pacers while the other 5 are spinners  
All 5 spinners are right arm spinners and 2 are leg spinners while 3 are off spinners  
All 5 pacers are right arm pacers
